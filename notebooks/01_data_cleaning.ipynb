# Data Cleaning - Online News Popularity

## Objective
Prepare the dataset for analysis by:
- Loading data
- Checking structure and quality
- Handling unnecessary columns
- Preparing features for SQL and ML

```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load Dataset
df = pd.read_csv("../data/raw/OnlineNewsPopularity.csv")

df.head()
```

```python
# Basic Info
df.shape
df.info()
```

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
# Drop unnecessary columns
df.drop(columns=["url", "timedelta"], inplace=True)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Create log-transformed target
df["log_shares"] = np.log1p(df["shares"])
```

```python
# Quick Summary Stats
df.describe()
```

```python
# Save cleaned data
df.to_csv("../data/processed/cleaned_online_news.csv", index=False)
```
